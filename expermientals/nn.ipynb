```python
#Load the json file
import json
import random
f = open("summaries.json")
rows = json.load(f)

img_1_summas = []
img_2_summas = []
model_summas = []

#Adding all data after basic normalization(divide by sum)

def basic_normalize(lst, dividend):
    temp = []
    for i in lst:
        temp.append(i/( dividend/ 1000))
    return temp

for row in rows:
    img_1_sum, img_2_sum, model_sum = sum(row["img_1_summa"]), sum(row["img_2_summa"]), sum(row["model_summa"])
    if img_1_sum >= 1000 and  img_2_sum >= 1000 and model_sum >= 1000:
        img_1_summas.append(basic_normalize(row["img_1_summa"], img_1_sum))
        img_2_summas.append(basic_normalize(row["img_2_summa"], img_2_sum))
        model_summas.append(basic_normalize(row["model_summa"], model_sum))
 

```
